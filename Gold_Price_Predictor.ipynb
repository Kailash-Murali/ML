Gold Price Predictor

Libraries

```python
import pandas as Pd
import numpy as Np
import seaborn as Sb
import matplotlib.pyplot as Mpl
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

Information about the data

```python
GoldData = Pd.read_csv('/content/gld_price_data.csv')
GoldData.head()
```

```python
GoldData.tail()
```

```python
GoldData.info()
```

```python
GoldData.describe()
```

```python
GoldData.isnull().sum()
```

Correlation

```python
GoldData['Date'] = Pd.to_datetime(GoldData['Date'])
Correlation = GoldData.corr()
```

```python
Sb.heatmap(Correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
print(GoldData.corr()['GLD'])
```

Features & Target

```python
X = GoldData.drop(['Date','GLD'],axis=1)
Y = GoldData['GLD']
```

```python
print(X, Y, sep = '\n')
```

Model Training

```python
XTrain, XTest, YTrain, YTest = train_test_split(X, Y, train_size = 0.2, random_state = 2)
Regressor = RandomForestRegressor(n_estimators = 100)
Regressor.fit(XTrain, YTrain)
TestDataPrediction = Regressor.predict(XTest)
print(TestDataPrediction)
```

Evaluation & Visualisation

```python
Error = metrics.r2_score(YTest, TestDataPrediction)
print("R Squared Error : ", Error)
```

```python
YTest = Np.array(YTest)
Mpl.plot(YTest, color='red', label = 'Actual Value')
Mpl.plot(TestDataPrediction, color='blue', label='Predicted Value')
Mpl.title('Actual Price vs Predicted Price')
Mpl.xlabel('Number of values')
Mpl.ylabel('GLD Price')
Mpl.legend()
Mpl.show()
```
